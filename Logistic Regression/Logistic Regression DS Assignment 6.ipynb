# Data Science Assignment 6 : Logistic Regression

#### 1. Find Wether the client has subscribed a term deposit or not Binomial ("yes" or "no")<br>

Attribute information For bank dataset<br>

   Input variables:<br>
   Bank client data:<br>
   1 - age (numeric)<br>
   2 - job : type of job (categorical: "admin.","unknown","unemployed","management","housemaid","entrepreneur","student",
                                       "blue-collar","self-employed","retired","technician","services")<br> 
   3 - marital : marital status (categorical: "married","divorced","single"; note: "divorced" means divorced or widowed)<br>
   4 - education (categorical: "unknown","secondary","primary","tertiary")<br>
   5 - default: has credit in default? (binary: "yes","no")<br>
   6 - balance: average yearly balance, in euros (numeric) <br>
   7 - housing: has housing loan? (binary: "yes","no")<br>
   8 - loan: has personal loan? (binary: "yes","no")<br>
   
   related with the last contact of the current campaign:<br>
   9 - contact: contact communication type (categorical: "unknown","telephone","cellular") <br>
  10 - day: last contact day of the month (numeric)<br>
  11 - month: last contact month of year (categorical: "jan", "feb", "mar", ..., "nov", "dec")<br>
  12 - duration: last contact duration, in seconds (numeric)<br>
   
   other attributes:<br>
  13 - campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)<br>
  14 - pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric, -1 means client was not previously contacted)<br>
  15 - previous: number of contacts performed before this campaign and for this client (numeric)<br>
  16 - poutcome: outcome of the previous marketing campaign (categorical: "unknown","other","failure","success")<br>

  Output variable (desired target):<br>
  17 - y - has the client subscribed a term deposit? (binary: "yes","no")<br>

  18. Missing Attribute Values: None<br>


```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"bank-full.csv",sep = ";")
df.head()
```

#### Exploratory Data Analysis

Dropping unwanted columns


```python
df.drop(["marital","month","poutcome"],axis=1,inplace=True)
```

LE.fit_transform() : Encode target labels with value between 0 and n_classes-1.

```python
from sklearn.preprocessing import LabelEncoder
LE= LabelEncoder()
df['job'] = LE.fit_transform(df['job'])
df['education'] = LE.fit_transform(df['education'])
df['default'] = LE.fit_transform(df['default'])
df['housing'] = LE.fit_transform(df['housing'])
df['loan'] = LE.fit_transform(df['loan'])
df['contact'] = LE.fit_transform(df['contact'])
df['y'] = LE.fit_transform(df['y'])
```

#### Model Building

Dividing our data into input and output variables 

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

Python provides pickle modules for Serialization and de-Serialization of python objects like lists, dictionaries, tuples, etc.<br>
Pickling is also called marshaling or flattening in other languages. <br>
Pickling is used to store python objects<br>


```python
from sklearn.linear_model import LogisticRegression
import pickle
```

Logistic regression and fit the model


```python
classifier = LogisticRegression()
classifier.fit(X,Y)
```

Saving the model to disk


```python
filename = "finalized_model.sav"
pickle.dump(classifier, open(filename,'wb'))
```

Predict for X dataset


```python
pickle.load(open(filename,'rb'))
```

Classifier.read_pickle_file('/content/finalized_model.sav')


```python
y_pred = classifier.predict(X)
```

```python
y_pred_df = pd.DataFrame({'actual':Y, 'predicted_prob':classifier.predict(X)})
```

```python
y_pred_df
```

#### Confusion Matrix for the model accuracy

It is a table that is used in classification problems to assess where errors in the model were made.<br>
The rows represent the actual classes the outcomes should have been. While the columns represent the predictions we have made.<br>
Using this table it is easy to see which predictions are wrong.


```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y,y_pred)
print(confusion_matrix)
```

#### Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(Y,y_pred))
```

#### ROC Curve

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
fpr,tpr,thresholds = roc_curve(Y, classifier.predict_proba(X)[:,1])
auc= roc_auc_score(Y,y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr,tpr,color = 'red', label='logit model(area = % 0.2f)' %auc)
plt.plot([0,1],[0,1], 'k--1')
plt.xlabel('False Positive Rate or [1- True Negative Rate]')
plt.ylabel("True Positive Rate")
plt.show()
```
