#### Hierarchical Clustering

Importing clusters libraries


```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

Normalization Function 

```python
def norm_func(i):
    x = (i - i.min())/(i.max()-i.min())
    return(x)
```

Normalize data frame (Considering the numerical part of data)


```python
df_norm = norm_func(df.iloc[:,1:])
df_norm
```

Create Dendrogram


```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(df_norm, method = 'complete'))
```

Create Clusters


```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'complete')
```

Model fitting on dataset


```python
hc.fit(df_norm)
```

Save cluster for chart


```python
y_hc = hc.fit_predict(df_norm)
y_hc
```

```python
df['h_clusterid'] = y_hc
df
```

```python
df1 = df.sort_values("h_clusterid")
df1.iloc[:,[0,-1]]
```

Extracting features from dataset


```python
x = df.iloc[:,-1]
y = df1.iloc[:, -2]
```

#### K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

Standardization of data


```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df.iloc[:,1:])
scaled_df
```

```python
kmeans = KMeans(n_clusters = 4, random_state = 0)
kmeans.fit(scaled_df)
kmeans.inertia_
```

```python
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters = i, random_state = 0)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
wcss
```

Build Cluster Algorithm


```python
cluster_new = KMeans(4, random_state = 42)
cluster_new.fit(scaled_df)
```

```python
cluster_new.labels_
```

Assign Cluster to the dataset


```python
df['clusterid_new'] = cluster_new.labels_
cluster_new.cluster_centers_ 
```

```python
df.groupby('clusterid_new').agg(['mean']).reset_index()
```

```python
df[df['clusterid_new'] == 3]
```

#### DBSCAN : Density Base Spatial Clustering of Applications with Noise

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
array = df.values
```

```python
st_scaler = StandardScaler()
X = st_scaler.fit_transform(array)
X
```

```python
dbscan = DBSCAN(eps = 5 , min_samples = 6)
dbscan.fit(X)
```

Noisy Samples are given the label: -1


```python
print(dbscan.labels_)
```

```python
cl = pd.DataFrame(dbscan.labels_, columns = ['clusters'])
```

```python
cl
```

```python
clustered = pd.concat([df,cl], axis =1)
clustered
```

-1 Shows outlier / extreme values for the dataset


```python
clustered[clustered['clusters'] == -1].head()
```

```python
clustered.mean()
```

```python
p = clustered.iloc[:,-1]
q = clustered.iloc[:,-2]
r = clustered.iloc[:,-3]
```

```python
plt.scatter(p,q)
plt.scatter(q,r)
plt.scatter(p,r)
plt.show()
```

```python
plt.plot(p,q)
plt.plot(q,r)
plt.plot(p,r)
plt.show()
```

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot(p,q,r)
plt.show()
```
